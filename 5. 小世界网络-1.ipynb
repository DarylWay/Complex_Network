# 1. K近邻规则网络的生成与可视化

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# 自己写一个规则网络的生成算法
def regular_graph(n, k):
    G = nx.Graph()
    nodes = list(range(n))# 节点标签设置为0到n-1
    # 每个节点与双侧各k//2个节点相连, 注意边的总数为n*k//2, 而非nk
    for i in range(1, k//2+1):# //表示向下取整的除法
        targets = nodes[i:]+nodes[0:i]
        G.add_edges_from(zip(nodes, targets))
    return G
```

```python
n = 20
k = 4
color_list = ['red', 'gray']
plt.figure(figsize=(5,4))
G = regular_graph(n, k)
pos = nx.circular_layout(G)
nx.draw(G, pos, node_size = 100, node_color = color_list[0], edge_color = color_list[1])
plt.title("K-regular", fontsize = 20)
```

# 2. WS小世界网络的生成与可视化

- 调用nx自带的watts_strogatz_graph函数, 当重连边概率为0时, 生成的是k近邻规则网络, 当p介于0与1之间时, 生成小世界网络, 当边的重连概率是1是, 说明所有边都发生了重连, 此时生成了随机网络

```python
plt.figure(figsize=(15,4))

# 绘制规则网络
plt.subplot(131)
p = 0
G1 = nx.watts_strogatz_graph(n, k, p)
pos1 = nx.circular_layout(G1)
nx.draw(G1, pos1, node_size = 100, node_color = color_list[0], edge_color = color_list[1])
plt.title("regular", fontsize = 20)

# 绘制WS小世界网络
plt.subplot(132)
p = 0.2
G2 = nx.watts_strogatz_graph(n, k, p)
pos2 = nx.circular_layout(G2)
nx.draw(G2, pos2, node_size = 100, node_color = color_list[0], edge_color = color_list[1])
plt.title("small-world", fontsize = 20)

# 绘制随机网络
plt.subplot(133)
p = 1.0
G3 = nx.watts_strogatz_graph(n, k, p)
pos3 = nx.circular_layout(G3)
nx.draw(G3, pos3, node_size = 100, node_color = color_list[0], edge_color = color_list[1])
plt.title("random", fontsize = 20)
```

# 3. NW小世界网络的生成与可视化

```python
plt.figure(figsize=(15,4))

# 绘制规则网络
plt.subplot(131)
p = 0
G1 = nx.newman_watts_strogatz_graph(n, k, p)
pos1 = nx.circular_layout(G1)
nx.draw(G1, pos1, node_size = 100, node_color = color_list[0], edge_color = color_list[1])
plt.title("regular", fontsize = 20)

# 绘制NW小世界网络
plt.subplot(132)
p = 0.2
G2 = nx.newman_watts_strogatz_graph(n, k, p)
pos2 = nx.circular_layout(G2)
nx.draw(G2, pos2, node_size = 100, node_color = color_list[0], edge_color = color_list[1])
plt.title("small-world", fontsize = 20)

# 绘制完全网络(相当于NW模型中取连边概率为1, 但是nw函数无法生成完全网络, 需要调用complete_graph函数)
plt.subplot(133)
G3 = nx.complete_graph(n)
pos3 = nx.circular_layout(G3)
nx.draw(G3, pos3, node_size = 100, node_color = color_list[0], edge_color = color_list[1])
plt.title("complete", fontsize = 20)
```

# 4. 小世界网络的度分布

```python
import numpy as np
```

```python
# 定义求度分布的函数
def get_pdf(G, k):
    N = len(G.nodes())
    
    Pk = []
    for ki in k:
        count = 0
        for node in G.nodes():
            if G.degree(node)==ki:
                count+=1
        Pk.append(count/N)
    return Pk
```

```python
# 以N=1000. K=6的WS模型的数值模拟结果为例
N = 1000
K = 6
samples = 100
p_rew = [0.1, 0.2, 0.4, 0.6, 1.0]
```

```python
plt.figure(figsize=(8, 6))
symbols = ["ro-", "bs-", "g*-", "yv-", "k^-"]
# 为了便于统计平均, 指定度分布区间[1, 16]
kmin, kmax = 1, 16
x = list(range(kmin, kmax+1))
c = 0
for p in p_rew:
    s = np.zeros(kmax-kmin+1)
    for i in range(samples):
        G = nx.watts_strogatz_graph(N, K, p)
        y = get_pdf(G, x)
        s += np.array(y)
        
    s = list(s/samples)
    # 剔除概率为0的点
    new_x = []
    new_y = []
    for i in range(len(x)):
        if s[i]!=0:
            new_x.append(x[i])
            new_y.append(s[i])
    
    plt.plot(new_x, new_y, symbols[c], label="$p_{rew}=$"+str(p))
    c+=1
plt.legend(loc = 0, fontsize = 14)
plt.xlabel("$k$")
plt.ylabel("$p_k$")
plt.yscale("log")
plt.xlim([kmin, kmax])
```
